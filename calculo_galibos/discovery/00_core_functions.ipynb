```python
from shapely.geometry import Polygon

# Define your pentagons as a list of points
pentagon1 = [(0,0), (1,2), (2,1), (1,0), (0,0)]
pentagon2 = [(2,1), (3,3), (4,2), (3,1), (2,1)]
pentagon3 = [(4,2), (5,4), (6,3), (5,2), (4,2)]
pentagon4 = [(6,3), (7,5), (8,4), (7,3), (6,3)]

# Create polygons
polygon1 = Polygon(pentagon1)
polygon2 = Polygon(pentagon2)
polygon3 = Polygon(pentagon3)
polygon4 = Polygon(pentagon4)

# Unify the polygons
unified = polygon1.union(polygon2).union(polygon3).union(polygon4)

# Extract the contour (boundary) of the unified shape
contour = unified.boundary

# Print the contour
print(contour)
```

```python
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# Define your pentagons as a list of points
pentagon1 = [(0,0), (1,2), (2,1), (1,0), (0,0)]
pentagon2 = [(2,1), (3,3), (4,2), (3,1), (2,1)]
pentagon3 = [(4,2), (5,4), (6,3), (5,2), (4,2)]
pentagon4 = [(6,3), (7,5), (8,4), (7,3), (6,3)]

# Create polygons
polygon1 = Polygon(pentagon1)
polygon2 = Polygon(pentagon2)
polygon3 = Polygon(pentagon3)
polygon4 = Polygon(pentagon4)

# Plot the individual pentagons
fig, ax = plt.subplots()
ax.plot(*polygon1.exterior.xy)
ax.plot(*polygon2.exterior.xy)
ax.plot(*polygon3.exterior.xy)
ax.plot(*polygon4.exterior.xy)
plt.show()

# Unify the polygons
unified = polygon1.union(polygon2).union(polygon3).union(polygon4)

# Plot the unified polygon
fig, ax = plt.subplots()
ax.plot(*unified.exterior.xy)
plt.show()
```
